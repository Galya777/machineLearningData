```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, OPTICS

# Зареждане на данни
data = pd.read_csv('basketball_performance.csv')
```

```python
# Проверка на първите редове на данните
print(data.head())

# Описание на статистическите характеристики
print(data.describe())

# Проверка за липсващи стойности
print(data.isnull().sum())

# Запълване на липсващите стойности
data.fillna(data.mean(), inplace=True)

# Проверка за екстремални стойности
plt.figure(figsize=(20,10))
sns.boxplot(data=data)
plt.xticks(rotation=90)
plt.show()
```

```python
# Корелационна матрица
correlation_matrix = data.corr()
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()

# Графики на разпределение на данните
plt.figure(figsize=(12, 6))
sns.histplot(data['PointsPerGame'], kde=True)
plt.title('Разпределение на средно количество отбелязани точки')
plt.show()

plt.figure(figsize=(12, 6))
sns.histplot(data['MinutesPlayed'], kde=True)
plt.title('Разпределение на общия брой изиграни минути')
plt.show()

# Scatter plot за връзка между изиграни минути и средно количество отбелязани точки
plt.figure(figsize=(12, 6))
sns.scatterplot(x='MinutesPlayed', y='PointsPerGame', data=data)
plt.title('Връзка между изиграни минути и средно количество отбелязани точки')
plt.show()
```

```python
# Стандартизиране на данните
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data.drop('Name', axis=1))

# Приложение на PCA
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

# Визуализация на резултатите от PCA
plt.figure(figsize=(12, 6))
plt.scatter(pca_data[:, 0], pca_data[:, 1])
plt.title('PCA резултати')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.show()
```

```python
# KMeans клъстеризация
kmeans = KMeans(n_clusters=2)
kmeans_labels = kmeans.fit_predict(pca_data)

plt.figure(figsize=(12, 6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=kmeans_labels, cmap='viridis')
plt.title('KMeans клъстеризация')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.show()

# Агломеративна клъстеризация
agg_clustering = AgglomerativeClustering(n_clusters=2)
agg_labels = agg_clustering.fit_predict(pca_data)

plt.figure(figsize=(12, 6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=agg_labels, cmap='viridis')
plt.title('Агломеративна клъстеризация')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.show()

# DBSCAN клъстеризация
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(pca_data)

plt.figure(figsize=(12, 6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=dbscan_labels, cmap='viridis')
plt.title('DBSCAN клъстеризация')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.show()

# OPTICS клъстеризация
optics = OPTICS(min_samples=5)
optics_labels = optics.fit_predict(pca_data)

plt.figure(figsize=(12, 6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=optics_labels, cmap='viridis')
plt.title('OPTICS клъстеризация')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.show()
```

```python
# KMeans с повече клъстери
kmeans_more_clusters = KMeans(n_clusters=4)
kmeans_more_labels = kmeans_more_clusters.fit_predict(pca_data)

plt.figure(figsize=(12, 6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=kmeans_more_labels, cmap='viridis')
plt.title('KMeans с повече клъстери')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.show()

# Анализ на резултатите преди и след скалиране
# KMeans преди скалиране
pca_data_unscaled = PCA(n_components=2).fit_transform(data.drop('Name', axis=1))

kmeans_unscaled = KMeans(n_clusters=2)
kmeans_labels_unscaled = kmeans_unscaled.fit_predict(pca_data_unscaled)

plt.figure(figsize=(12, 6))
plt.scatter(pca_data_unscaled[:, 0], pca_data_unscaled[:, 1], c=kmeans_labels_unscaled, cmap='viridis')
plt.title('KMeans преди скалиране')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.show()
```
